### Figure 1

```python
data=iris
seaborn=sns.load_dataset("iris")
sns.histplot (data = iris, x = "SepalWidthCm", hue = "Species", multiple = "stack")
```

This is a stacked bar graph that reflects the amount and the Sepal width of the species reported in the table in Task 1. The x-axis refers to the Sepal widths of the species in cm, the y-axis refers to the amount of each species there are, and the colors (blue, orange, and green) labels what kind of species are being reported on. 

### Figure 2

```python
sns.kdeplot(data = iris, x = "PetalLengthCm", hue = "Species", multiple = "stack")
```

This is a area chart that reflects the densities and the Petal lengths of the species reported in the table in Task 1. The x-axis refers to the Petal lengths of the species in cm, the y-axis refers to their densities, and the colors (blue, orange, and green) labels what kind of species are being reported on. 

### Figure 3

```python
sns.pairplot(data = iris, hue = "Species")
```

This is a 4x4 plot that reflects pairwise relationships. Each variable is shared in the x-axis across the row and y-axis across the column. The x and y axis records the Petal lengths and widths, as well as the Sepal lengths and widths (all in cm) of the species. The colors (blue, orange, and green) labels what kind of species are being reported on.
