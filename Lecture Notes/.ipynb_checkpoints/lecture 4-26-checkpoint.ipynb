Basic usage of pyplot: Very similiar to Matlab

```python
# matlab way
import numpy as np
x = np.linspace(0, 10, 100)
# initialize the figure below; this isn't required for the code to work but it's good for future use
fig = plt.figure(figsize=(8, 6),dpi=220) # create the figure, just like figure() in matlab
# dpi is sort of like resolution, the higher it is the higher resolution the picture is
# figure is a class, and plt is the submodule of matplotlib that we imported
plt.plot(x, np.sin(x), linestyle = '-',color = 'b',label='sin')  # label is used for legend
# this is for the sin graph, it shows the different adjustments you can make to your graph; this is used as legend of graph
plt.plot(x, np.cos(x), '--g', label = 'cos')
# this is for the cos graph, this is a shorter way you can make your legend
plt.xlim(-1, 11)
# domain of the graph
plt.title("A Sine Curve") 
plt.xlabel('x') 
plt.ylabel("sin(x)")
plt.legend()
```

The object-oriented feature is more evident in making subplots. [Explore more usages here](https://matplotlib.org/3.1.0/gallery/subplots_axes_and_figures/subplots_demo.html).

```python
# subplots
fig, ax = plt.subplots(2, dpi =150) # subplots(n,m); n represents number of rows your subplot lays as; m represents columns
# subplots (2) automatically creates two subplots, one on top of the other
# subplots (1,2) will give two subplots side by side
# subplots help you create figures as a whole
# ax stands for axes
# 2 stands for vector, first element is first subplot, it's defined by ax[0] in the following commands
# second elements is second subplot, it's defined by ax[1] in the following commands
ax[0].plot(x, np.sin(x)) # plot and set_title are the methods of ax[0] -axes
ax[0].set_title('sin')
# ax is an subplot; ax[0] indicates what you want to do with the first subplot
# plot and set title are different function / methods that can be used with the ax
ax[1].plot(x, np.cos(x))
ax[1].set_title('cos')
```

## Image Processing

There are many great packages available to handle the image data in Python, such as [Pillow](https://pillow.readthedocs.io/en/stable/handbook/tutorial.html#using-the-image-class), [Scikit-Image](https://scikit-image.org/) and [opencv-python](https://github.com/skvark/opencv-python).

Here we import images from Scikit-Image which is [well-compatible with Numpy](https://scikit-image.org/docs/dev/user_guide/numpy_images.html), and use Numpy to manipulate images.

**Representing an image**

*Grayscale*
- imagine an image as a matrix, representing pixels. Each pixel (representing an element of the matrix has) 1 number, ranging from 0 to 255
- white: closer to 255; black: closer to 0
    - dimension: m x n

*Colored*
- 3 matrix; called rgb
- rgb: red green blue; each color can be decomposed into these three color channels
- each pixel correspond to these three color channels, each color image has three matrices associated with it
    - each element correspond to the following pixels
    - tensor: three matrices stacked onto one another
    - dimension: m x n x 3

```python
from skimage import data # imagine data as some kind of submodule containing some functions
image_astro = data.astronaut() # read the image as numpy array
image_rock = data.rocket() 
fig = plt.figure(dpi=100) # initialize new figure
plt.imshow(image_astro) # image show (imshow)
```

```python
fig = plt.figure(dpi=100)
plt.imshow(image_rock)
# note: the pixels underneath the images serve as pixel indexes (y axis: m); (x-axis: n)
# imagine that this image is made up of very small color blocks that come together
```

In data science, a common way to store image is through 2D matrix (gray) or 3D tensor (RGB color).

For instance, a gray-scale image with size $m\times n$ can be represented by a matrix $I_{1}\in\mathbb{R}^{m\times n}$, whose elements denotes the intensities of pixels. 

A color image $m\times n$ can be represented by a tensor (or you can imagine three matrices stacked together) $I_{2}\in\mathbb{R}^{m\times n\times 3}$, where the three $m\times n$ matrices denote the intensity in red, green and blue channels respectively (basic assumption is any color can be decomposed in RGB)

Even with simple Numpy expressions, you can do some image processing like in Photoshop!

- Crop the images

```python
# slicing of the image
image_astro_split = image_astro[:427,:,:] # crop it into lowest dimension to match with the dimension of the rocket image
# cuts the bottom of the image
image_rock_split = image_rock[:,:512,:] # crop it into 512 to match with dimension of astronaut; cuts the side of the image
```

```python
image_rock_split.shape
# shape is an attribute used to check the dimensions of a certain image
```

```python
fig, ax = plt.subplots(ncols=2, dpi = 100)
ax[0].imshow(image_astro_split) # plot and set_title are the methods of ax[0] -axes
ax[1].imshow(image_rock_split)
```

- Invert the color intensities

```python
fig = plt.figure(dpi=100)
plt.imshow(255-image_rock) # changes color intensity; something light becomes dark and something dark becomes light
# inversion of the image
```

- Exchange RGB channels

```python
fig = plt.figure(dpi=100)
plt.imshow(image_rock[:,:,[0,2,1]]) # we keep the 1st and 2nd dimensions of the numpy array
# fancy index: this makes rgb become rbg; 0 means red stays the same, 2 is b, and 1 is g
```

```python
fig = plt.figure(dpi=100)
plt.imshow(image_rock[:,:,[2,1,0]])
# rgb becomes bgr; 2 is b, 1 is g, and 0 is r
```

- Binarize the image

```python
# normally, images are smooth and colors blend in very smoothly
# binarizing an image makes the colors stand out and shows what main colors are a part of that blend
image = image_rock
image_bi = np.empty_like(image) # creates an empty numpy array

thresh = 90
maxval = 255

for i in range(3): #loop over each color channel
    image_bi[:, :, i] = (image[:, :, i] > thresh) * maxval # always pick up the last channel
    # if it is greater than the threshold, then you multiply it with 255, if not, you make it zero

fig = plt.figure(dpi=100)
plt.imshow(image_bi[:,:,])
```

```python
image_bi = np.empty_like(image) # creates an empty numpy array

thresh = 90
maxval = 255

for i in range(3): #loop over each color channel
    image_bi[:, :, i] = (image[:, :, i] > thresh) * maxval 
    
fig = plt.figure(dpi=100)
plt.imshow(image_bi[:,:,[2,1,0]]) # exchange the colors
```

- Blending

```python
# blend both images together
image_combine = 0.2*image_astro_split+0.5*image_rock_split # each pixel is 0.4 of the astronaut and 0.6 of the rocket
# if we increase astronaut and decrease rocket, we get a picture with the astronaut being more prominent
fig = plt.figure(dpi=100)
plt.imshow(image_combine.astype('uint8')) # change type to integer; to use image show
plt.axis('off')
```

### Convolution
Imagine we have an image and a kernel, each represented by a 3x3 matrix;
I represents the image and K represents the kernel, and C is the convolution.

`C[1,1]=I[0,0]K[0,0]+I[0,1]K[0,1]+...+I[1,0]K[1,0]+I[1,1]K[1,1]+...+I[2,2]K[2,2]`

Note: there are 9 numbers within the image matrix and 9 in the kernel matrix

This method is messy, as there are no pixels on the outer edge

Solution: Create a (m+2)x(n+2) matrix (5x5 in this case), where the outer edge is zero and the inner matrix has the same dimensions as the image and kernel matrices

Let I~ be the new image, so

`C[1,1]=I~[1,1]K[0,0}+...=sum(I~[1:4,1:4]*K)`

So, the general function of convolution is

`C[i,j]=sum(I~[i:i+3,i:j+3]*k)`
Here, 3 is for 3x3 matrix
